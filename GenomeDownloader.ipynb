```python
for idx, row in tqdm(df.iterrows(), total=len(df)):
  strain_id = str(row['ID_strains']).strip()
  taxid = str(row['NCBI_taxid']).strip()

  zip_file = "ncbi_dataset.zip"
  rename = f"{strain_id}.zip"

  out_path = os.path.join(output_folder, rename)

  if os.path.exists(out_path):
    print(f"Skipping {strain_id}. Already exists")
    continue

  try:
    print(f"Downloading TaxID {taxid} as {strain_id}")
    subprocess.run(["datasets", "download", "genome", "taxon", taxid], check=True)
    shutil.move(zip_file, out_path)
  except subprocess.CalledProcessError as e:
    print(f"[!] Error. Failed to download TaxID {taxid}: {e}. Skipping")
    continue
  except Exception as e:
    print(f"[!] Unexpected Error with TaxID {taxid}: {e}. Skipping")
    continue
```
